# Méthode de dichotomie

Appliqueé à la fonction f: $x\longmapsto e^{\frac{-x^2}{2}}-x$ sur l'intervalle $[0,1]$

```python
import numpy as np 
from matplotlib import pyplot as plt,animation as anim
```

```python
f=lambda x:np.exp(-x**2/2)-x
x=np.linspace(0,1,100)
plt.plot(x,f(x))
plt.plot(x,[0]*len(x))
```

```python
def dichotomie(f,a,b):
    x=a
    y=b
    while True:
        mid=(x+y)/2
        yield mid
        if(f(mid)==0):
            break
        if f(x)*f(mid)<=0:
            y=mid
        else:
            x=mid
```

```python
(a,b)=(0,1)
dico=dichotomie(f,a,b)
alpha=0
try :
    
    for i in range(100):
        alpha=next(dico)
        print(str(i)+"  "+str(alpha))
except:
    pass
```

```python
f(alpha)
```

```python
plt.plot(x,f(x))
plt.plot(x,[0]*len(x))
plt.plot(alpha,f(alpha),'bo')
```

## Méthode de point fixe

Appliqueé à la fonction g: $x\longmapsto \exp(-\frac{x^2}{2})$ sur l'intervalle $[0,1]$

$$ |g'(x)|=|-x\exp(-\frac{x^2}{2})|\leq e^{-\frac{1}{2}}<1 \quad \forall x \in [0,1] $$

```python
g=lambda x:f(x)+x
#gprime=lambda x:x*np.exp(-x**2/2)
```

```python
plt.plot(x,g(x))
#plt.plot(x,gprime(x))
plt.plot(x,x)
```

```python
def point_fixe(g, x0):
    x1 = x0
    while True:
        x1 = g(x1)
        yield x1
        if x1==g(x1):
            break
```

```python
xn=point_fixe(g,0)
try :
    for i in range(100):
        alpha=next(xn)
        print("{}  {}".format(i,alpha))
except:
    pass
```

```python
f(alpha)
```

```python
g(alpha)==alpha
```

```python
plt.plot(x,g(x))
#plt.plot(x,gprime(x))
plt.plot(x,x)
plt.plot(alpha,alpha,'bo')
```

# méthode de la sécante

```python
plt.plot(x,np.zeros(len(x)))
plt.plot(x,f(x))
```

```python
def secante(f,a,b):
    x0=a
    x1=b
    while True:
        x2=(x1*f(x0)-x0*f(x1))/(f(x0)-f(x1))
        yield x2
        if f(x2)*(x0)<=0:
            x1=x2
        else:
            x0=x2
        if f(x2)==0:
            break
```

```python
xnn=secante(f,a,b)
try:
    for i in range(40):
        alpha_sec=next(xnn)
        print(str(i)+"  "+str(alpha_sec))
except:
    pass
```

```python
f(alpha_sec)
```

# Méthode de newthon

```python
def newton(f,fp,x0):
    g=lambda x:x-f(x)/fp(x)
    ptfixegen=point_fixe(g,x0)
    while True:
        yield next(ptfixegen)
```

```python
fp=lambda x:-x*np.exp(-x**2/2)-1
```

```python
ntn=newton(f,fp,0)
try:
    for i in range(40):
        alpha_nt=next(ntn)
        print(str(i)+"  "+str(alpha_nt))
except:
    pass
```
